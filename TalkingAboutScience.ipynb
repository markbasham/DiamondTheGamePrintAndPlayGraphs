```python
import pandas as pd
df = pd.read_csv("data/TalkingAboutScience.csv")
print(df)
```

```python
import matplotlib.pylab as pl
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.dpi"] = 300
title = "Most used 'being a scientist means...' responses related to how\noften people talk about science"
ax = df.plot(kind='bar', x='Word', rot=90, stacked=True, colormap=pl.cm.viridis, figsize=(8, 6), title=title)
ax.set_ylabel("Number of People")
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('figures/TalkingAboutScience.png', dpi=300)
```
