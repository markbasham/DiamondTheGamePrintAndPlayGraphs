```python
import pandas as pd
# Load Print and play data
qrdf = pd.read_csv("data/Questionaire_Responces.csv")

qrdf['Q2'][qrdf['UK'] == 1.0] = 'UK'
qrdf['Q2'][qrdf['Q2'] == 'United States of America'] = 'USA'
qrdf['Q2'][qrdf['Q2'] == 'US'] = 'USA'

del qrdf['UserID']
del qrdf['UserNo']
del qrdf['Started']
del qrdf['Ended']
del qrdf['Q1']
del qrdf['UK']
del qrdf['Q5']
del qrdf['Q6']
qrdf['Q3.1'] = pd.to_numeric(qrdf['Q3.1. 0 - 7 years old'], errors='coerce').fillna(0)
qrdf['Q3.2'] = pd.to_numeric(qrdf['Q3.2. 8 - 14 years old'], errors='coerce').fillna(0)
qrdf['Q3.3'] = pd.to_numeric(qrdf['Q3.3. 15 - 19 years old'], errors='coerce').fillna(0)
qrdf['Q3.4'] = pd.to_numeric(qrdf['Q3.4. 20 +'], errors='coerce').fillna(0)

qrdf['participants'] = qrdf['Q3.1'] + qrdf['Q3.2'] + qrdf['Q3.3'] + qrdf['Q3.4'] 



Top5 = qrdf.groupby(by='Q2').sum().sort_values(by=['participants'], ascending=False)[0:23]
print(Top5)
Top5.plot(kind='bar', y='participants', logy=False)
```

```python
import pandas as pd
df = pd.read_csv("data/TalkingAboutScienceByCountry.csv")
df['total'] = df['Everyday'] + df['Once or twice a week'] + df['Once or twice a month'] + df['Once or twice a year'] + df['Never']
df['Everyday'] /= df['total']/100.
df['Once or twice a week'] /= df['total']/100.
df['Once or twice a month'] /= df['total']/100.
df['Once or twice a year'] /= df['total']/100.
df['Never'] /= df['total']/100.
df['Country'] += df['total'].apply('\n({} respondants)'.format)
del df['total']
print(df)
```

```python
import matplotlib.pylab as pl
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.dpi"] = 300
title = "Responses to how often people talk about science in the 5 Countries with most players"
ax = df.plot(kind='bar', x='Country', rot=0, stacked=True, colormap=pl.cm.viridis, figsize=(8, 6), title=title)
ax.set_ylabel("Percentage of respondents answering")
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('figures/TalkingAboutScienceByCountry.png', dpi=300)
```
